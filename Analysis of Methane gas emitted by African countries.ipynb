```python
df = df.drop(['notes','Unnamed: 0','region'],axis = 1)
```

```python
df.head()
```

 # Scatter plot

```python
fig = px.scatter(df, x="baseYear",y="emissions")
fig.show()
```

# Bar chart

```python
fig = px.bar(df, x=df.country, y=df.emissions,color="type", animation_frame='baseYear',height=1000)
fig.show()
```

fig = px.bar(df, x=df.Location, y=df.Diagnosis, hover_name="Sex",color=df.Sex, height=500)
fig.show()

# Funnel

```python
px.funnel(df,x='BP Systolic',y='Age',color=df.County, hover_name="Location",height=700)
```

# Line

```python
px.line(df,x="BP Systolic",y="BMI",hover_name="Sex", color=df.Diagnosis,height=700)
```

# Histogram

```python
px.histogram(df,x=['Weight'],y='BMI',color=df.Location, marginal='rug', hover_name="Sub County", height=700)
```

# Sunburst

```python
px.sunburst(df,path=["Sub County",'Location'], values="Frequency of clinic visit", hover_name="Age", color="Education", height=700)
```

```python
px.sunburst(df,path=["County",'Location'], values="Age", hover_name="Diagnosis", color="Location", height=700)
```

# Treemap

```python
px.treemap(df,path=["County",'Location'], values="Frequency of clinic visit", hover_name="Diagnosis", color="Sex", height=700)
```

```python
px.ecdf(df, x="Age",y="BP Systolic",hover_name="Sex",hover_data=df.columns,color="Diagnosis", height=700)
```

# Violin

```python
px.violin(df, x="Age",y="Sex",hover_name="Weight",hover_data=df.columns,color="Diagnosis", height=700)
```
