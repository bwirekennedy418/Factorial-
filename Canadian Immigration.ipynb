```python
#lets create a dataframe for three countries
df1= df_can.iloc[[56, 67, 71],:]
df1
```

step1 of creating a waffle chart is by determining the proportion of each category with respect to total

```python
total_values= df1['Total'].sum()
category_proportion= df1['Total']/ total_values
category_proportion
```

step2 is defining the overall size of the waffle chart

```python
width=15
height=15
total_num_tiles= width * height
print(f' Total number of tiles is {total_num_tiles}.')
```

step 3 Using the proportion of each category o determine the respective number of tiles

```python
Tiles_per_category= (category_proportion * total_num_tiles).round().astype(int)
Tiles_per_category
```

step 4  creating a matrix that resembles the waffle chart and populating it

```python
#initialize the waffle chart as an empty matrix
waffle_chart= np.zeros((width, height), dtype= np.uint)

#define indices to loop through the waffle chart
category_index=0
tile_index=0

#populate the waffle chart
for col in range(width):
    for row in range(height):
        tile_index +=1
        #if the number of tiles populated for the current category is equl to its corresponding allocated tiles
        if tile_index > sum(Tiles_per_category[0: category_index]):
            # proceed to the next category
            category_index +=1
            #set the class value to an integer, which increase the class
            waffle_chart[row, col] = category_index
            print("Waffle chart populated")
```

```python
waffle_chart
```

step 5 Map the waffle chart matrix into visual

```python
#instantiate the new figure object
fig= plt.figure
#use matshow to display the waffle chart
colormap= plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()
plt.show()

#get the axis
ax= plt.gca()

ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
ax.set_yticks(np.arange(-.5, (height), 1), minor=True)

#add gridlines based on minor ticks
ax.grid(which='minor', color='green', linestyle='dashed', linewidth=2)

plt.xticks([])
plt.yticks([])
plt.show()
```
